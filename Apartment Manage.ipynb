```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
import ipywidgets as widgets
from IPython.display import display, HTML, Image, clear_output
from ipywidgets import Layout
import ipywidgets as widgets
from google.colab import files
from PIL import Image as PILImage

# Định nghĩa cấu trúc dữ liệu và file lưu trữ
headers = ['ID_apartment', 'Area', 'Floor', 'Price', 'Status', 'Num_of_rentals']
headers_customer = ['ID', 'First_name', 'Last_name', 'Age', 'Gender', 'ID_apartment']
headers_attachment = ['ID_attachment', 'Image_Path', 'ID_apartment']
drive.mount('/content/drive')
csv_path = '/content/drive/MyDrive/CSV/Room.csv'
csv_pathCus = '/content/drive/MyDrive/CSV/Customer_ap.csv'
csv_pathAttachment = '/content/drive/MyDrive/CSV/attachment.csv'
rooms_db = csv_path
cus_db = csv_pathCus
attachment_db = csv_pathAttachment

# label
label_room = widgets.HTML(value='<div style="font-size: 24px; color: #3498db; font-weight: bold; margin-left: 100px; text-transform: uppercase";">Quản lý phòng</div>')
label_cus = widgets.HTML(value='<div style="font-size: 24px; color: #3498db; font-weight: bold; margin-left: 100px; margin-top: 40px; text-transform: uppercase";">Quản lý khách</div>')
label_item = widgets.HTML(value='<div style="font-size: 12px; color: red; font-weight: bold;"></div>')
label_item2 = widgets.HTML(value='<div style="font-size: 12px; color: red; font-weight: bold;"></div>')
# Creat book widgets
id_a = widgets.Text(description ='ID phòng:')
area = widgets.Text(description ='Kích thước:')
floor = widgets.Text(description = 'Tầng:')
status_options = ['Empty', 'For-rent']
status = widgets.Dropdown(description='Trạng thái:', options=status_options)
price = widgets.Text(description = 'Giá:')
nor = widgets.Text(description = 'Lượt thuê:')
add_button = widgets.Button(description ='Xác nhận', button_style ='success')
add_button_open = widgets.Button(description ='Thêm Phòng', button_style ='success')

# Thêm widget cho việc chọn nhiều hình ảnh
upload_button = widgets.FileUpload(accept='.jpg, .png', multiple=True)

# Delete widgets
rooms_df = pd.read_csv(rooms_db)
room_options_all = sorted(list(rooms_df['ID_apartment']))
delete_id_a = widgets.Dropdown(description='ID phòng:', options=room_options_all)
delete_button = widgets.Button(description ='Xóa', button_style ='danger')
delete_button_rent = widgets.Button(description ='Vẫn xóa', button_style ='danger')
delete_button_cancel = widgets.Button(description ='Hủy', button_style ='danger')
delete_button_open = widgets.Button(description ='Xóa phòng', button_style ='danger')
# Update widgets
id_all_room = widgets.Dropdown(description='ID phòng:', options=room_options_all)
update_area = widgets.Text(description='Diện tích:')
update_floor = widgets.Text(description='Tầng:')
update_price = widgets.Text(description='Giá:')
update_button = widgets.Button(description ='Xác nhận', button_style ='success')
update_button_open = widgets.Button(description ='Cập nhật phòng', button_style ='success')
# Trường nhập và nút chi tiết
detail_button = widgets.Button(description='Chi tiết', button_style='info')
# List book widgets
list_button = widgets.Button(description ='Liệt kê phòng', button_style ='primary')
# Chart widgets
chart_button = widgets.Button(description ='Hiển thị biểu đồ', button_style ='info')
status_button_pie_chart = widgets.Button(description='Hiển thị biểu đồ trạng thái phòng', button_style='info')
list_rooms_by_floor_pie_chart_button = widgets.Button(description='Liệt kê số phòng theo tầng (Biểu đồ tròn)', button_style='primary')
# List empty rooms button
list_empty_button = widgets.Button(description='Liệt kê phòng trống', button_style='success')

# Widgets cho nút thuê phòng
rooms_df = pd.read_csv(rooms_db)
for_rent_rooms = rooms_df[rooms_df['Status'] == 'Empty']
room_options_empty = sorted(list(for_rent_rooms['ID_apartment']))
rent_id_a = widgets.Dropdown(description='ID phòng:', options=room_options_empty)
rent_button = widgets.Button(description='Thuê phòng', button_style='success')
rent_button_confirm = widgets.Button(description='Xác nhận', button_style='success')

# Widgets cho thông tin khách hàng
customer_id = widgets.Text(description='ID:')
first_name = widgets.Text(description='Họ:')
last_name = widgets.Text(description='Tên:')
age = widgets.IntText(description='Tuổi:')
gender_options = ['Nam', 'Nữ', 'Khác']
gender = widgets.Dropdown(description='Giới tính:', options=gender_options)

# Search customer widgets
rooms_df = pd.read_csv(rooms_db)
for_rent_rooms = rooms_df[rooms_df['Status'] == 'For-rent']
room_options = sorted(list(for_rent_rooms['ID_apartment']))
search_customer_room = widgets.Dropdown(description='ID phòng:', options=room_options)
search_customer_name = widgets.Text(description='Tên khách hàng:')
search_customer_button = widgets.Button(description='Tìm kiếm khách hàng', button_style='primary')

# Widgets cho nút thêm khách hàng
add_customer_button = widgets.Button(description='Xác nhận', button_style='success')
add_customer_button_open = widgets.Button(description='Thêm khách hàng', button_style='success')

# Widgets cho nút trả phòng
return_id_a = widgets.Dropdown(description='ID phòng:', options=room_options)
return_button = widgets.Button(description='Trả phòng', button_style='danger')

# search
search_options = ['Chọn phương thức tìm kiếm','ID_apartment', 'Floor','Price', 'Status']
# Tạo Dropdown
search_drop = widgets.Dropdown(description='Tìm kiếm theo:', options=search_options)
search = widgets.Text(description = 'Giá:')
search_button = widgets.Button(description ='Tìm kiếm', button_style ='success')
status_input = widgets.Dropdown(description='Trạng thái:', options=status_options)
# options tầng lấy từ csv
rooms_df = pd.read_csv(rooms_db)
floor_options = sorted(rooms_df['Floor'].unique())

# out cus
output_search = widgets.Output()


# Tạo output widget để hiển thị thông báo
output = widgets.Output()
output = widgets.Output(layout={'border': '1px solid black', 'height': '500px', 'width': '800px', 'overflow_x': 'scroll'})
# out cus
output_cus = widgets.Output()
output_cus = widgets.Output(layout={'border': '1px solid black', 'height': '500px','width': '800px', 'overflow_x': 'scroll'})


# Dropdown list for floors
floor_dropdown = widgets.Dropdown(description='Chọn tầng:', options=floor_options)


# Function to update the chart based on the selected floor
def update_chart(change):
    # Filter data based on the selected floor
    floor_selected = change['new']
    filtered_data = rooms_df[rooms_df['Floor'] == floor_selected]

    # Create data for the pie chart
    price_ranges = ['1.000.000 VNĐ - 2.000.000 VNĐ', '2.000.000 VNĐ - 3.000.000 VNĐ']
    counts = [
        len(filtered_data[(filtered_data['Price'] >= 1000000) & (filtered_data['Price'] <= 2000000)]),
        len(filtered_data[(filtered_data['Price'] > 2000000) & (filtered_data['Price'] <= 3000000)])
    ]

    # Create a smaller pie chart
    with output:
        clear_output(wait=True)
        plt.figure(figsize=(6, 6))  # Adjust the figsize here
        plt.pie(counts, labels=price_ranges, autopct=lambda p: '{:.0f}'.format(p * sum(counts) / 100), startangle=140, colors=['blue', 'green'])
        plt.title(f'Số lượng phòng theo giá của tầng {floor_selected}')
        plt.show()


# Định nghĩa hàm xử lý sự kiện onChange
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        new_status = change['new']
        if new_status == 'Price':
          with output_search:
            clear_output(wait=True)
            button_layout_search = widgets.HBox([search, search_button], layout=widgets.Layout(margin='10px'))
            display(button_layout_search)
        elif new_status == 'Status':
          with output_search:
            clear_output(wait=True)
            status_input.options = status_options
            status_input.description = 'Trạng thái'
            button_layout_search = widgets.HBox([status_input, search_button], layout=widgets.Layout(margin='10px'))
            display(button_layout_search)
        elif new_status == 'Floor':
          with output_search:
            clear_output(wait=True)
            status_input.options = floor_options
            status_input.description = 'Tầng'
            button_layout_search = widgets.HBox([status_input, search_button], layout=widgets.Layout(margin='10px'))
            display(button_layout_search)
        elif new_status == 'ID_apartment':
          with output_search:
            clear_output(wait=True)
            status_input.options = room_options_all
            status_input.description = 'Mã phòng'
            button_layout_search = widgets.HBox([status_input, search_button], layout=widgets.Layout(margin='10px'))
            display(button_layout_search)

def handle_search_button_click(b):
    search_option = search_drop.value
    search_value_input = search.value.strip()
    search_value = status_input.value

    with output_search:
        clear_output(wait=True)

        if search_option == 'Price':
            if pd.api.types.is_numeric_dtype(rooms_df['Price']):
                result_df = rooms_df[rooms_df['Price'] == float(search_value_input)]
            else:
                rooms_df['Price'] = rooms_df['Price'].str.extract('(\d+)').astype(float)
                result_df = rooms_df[rooms_df['Price'] == float(search_value_input)]

        elif search_option == 'Status':
            result_df = rooms_df[rooms_df['Status'] == search_value]

        elif search_option == 'Floor':
            result_df = rooms_df[rooms_df['Floor'] == int(search_value)]

        elif search_option == 'ID_apartment':
            result_df = rooms_df[rooms_df['ID_apartment'] == search_value]

        with output:
            clear_output(wait=True)

            if result_df.empty:
                print("No results found.")
            else:
                display(result_df)

# Hàm để tính toán số lượng phòng theo tầng
def calculate_rooms_by_floor():
    rooms_df = pd.read_csv(rooms_db)
    rooms_by_floor = rooms_df.groupby('Floor').size().reset_index(name='Number_of_rooms')
    return rooms_by_floor

# Hàm để cập nhật biểu đồ tròn và hiển thị
def update_rooms_by_floor_pie_chart(button):
    rooms_by_floor_df = calculate_rooms_by_floor()

    # Tạo dữ liệu cho biểu đồ tròn
    labels = rooms_by_floor_df['Floor']
    sizes = rooms_by_floor_df['Number_of_rooms']
    colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6', '#34495e']

    # Tạo đối tượng Figure mà không hiển thị trực tiếp
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    ax.set_title('Số lượng phòng theo tầng')

    # Hiển thị biểu đồ trong widget Output
    with output:
        clear_output(wait=True)
        display(fig)

    # Đóng biểu đồ sau khi đã hiển thị trong widget Output
    plt.close(fig)

# Hàm để tính toán số lượng phòng trống và phòng đang cho thuê
def calculate_room_status():
    rooms_df = pd.read_csv(rooms_db)
    empty_rooms_count = len(rooms_df[rooms_df['Status'] == 'Empty'])
    rented_rooms_count = len(rooms_df[rooms_df['Status'] == 'For-rent'])
    return empty_rooms_count, rented_rooms_count

# Hàm để cập nhật biểu đồ tròn
def update_room_status_pie_chart(button):
    empty_count, rented_count = calculate_room_status()

    # Tạo dữ liệu cho biểu đồ tròn
    labels = ['Phòng trống', 'Phòng đang cho thuê']
    sizes = [empty_count, rented_count]
    colors = ['#3498db', '#e74c3c']

    # Tạo đối tượng Figure mà không hiển thị trực tiếp
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.pie(sizes, labels=labels, colors=colors, autopct=lambda p: '{:.0f}'.format(p * sum(sizes) / 100), startangle=140)
    ax.set_title('Trạng thái phòng')

    # Hiển thị biểu đồ trong widget Output
    with output:
        clear_output(wait=True)
        display(fig)

    # Đóng biểu đồ sau khi đã hiển thị trong widget Output
    plt.close(fig)

# Hàm xử lý việc thuê phòng
def rent_room(b):
    room_id_value = rent_id_a.value.strip() if rent_id_a.value else ''

    # Kiểm tra xem Mã phòng có hợp lệ không
    rooms_df = pd.read_csv(rooms_db)
    if room_id_value not in rooms_df['ID_apartment'].values:
        with output_cus:
            clear_output(wait=True)
            print(f"Phòng '{room_id_value}' không tồn tại.")
    else:
        # Hiển thị các trường nhập thông tin khách hàng
        label_item.value = f'<div style="font-size: 14px; color: red; font-weight: bold;">Thông tin thuê phòng:{room_id_value}</div>'
        with output_cus:
            clear_output(wait=True)
            display(label_item, customer_id, first_name, last_name, age, gender)
            display(rent_button_confirm)

# Xác nhận thêm khách hàng
def confirm_add_customer(b):
    # Lấy thông tin khách hàng
    customer_id_value = customer_id.value.strip()
    first_name_value = first_name.value.strip()
    last_name_value = last_name.value.strip()
    age_value = age.value
    gender_value = gender.value
    room_id_value = search_customer_room.value.strip()

    # Đọc DataFrame từ tệp phòng
    rooms_df = pd.read_csv(rooms_db)

    # Kiểm tra xem Mã phòng có hợp lệ không
    if room_id_value not in rooms_df['ID_apartment'].values:
        with output_cus:
            clear_output(wait=True)
            print(f"Phòng '{room_id_value}' không tồn tại.")
    else:
        # Thêm thông tin khách hàng vào tệp CSV khách hàng
        customer_data = {'ID': customer_id_value, 'First_name': first_name_value, 'Last_name': last_name_value,
                         'Age': age_value, 'Gender': gender_value, 'ID_apartment': room_id_value}
        customer_df = pd.read_csv(cus_db)
        customer_df = customer_df.append(customer_data, ignore_index=True)
        customer_df.to_csv(cus_db, index=False)

        # Cập nhật trạng thái của phòng thành 'For-rent'
        rooms_df.loc[rooms_df['ID_apartment'] == room_id_value, 'Status'] = 'For-rent'
        update_room_options()
        update_room_options_empty()
        # Hiển thị thông báo xác nhận thuê phòng
        with output_cus:
            clear_output(wait=True)
            print(f"Đã thuê phòng cho khách hàng {first_name_value} {last_name_value} (ID: {customer_id_value}).")
    # Xóa nội dung các widgets sau khi thuê phòng
    clear_customer_inputs()

# Hàm xử lý xác nhận thuê phòng
def confirm_rent(b):
    # Lấy thông tin khách hàng
    customer_id_value = customer_id.value.strip()
    first_name_value = first_name.value.strip()
    last_name_value = last_name.value.strip()
    age_value = age.value
    gender_value = gender.value

    # Lấy mã phòng đã chọn
    room_id_value = rent_id_a.value.strip()

    # Đọc DataFrame từ tệp phòng
    rooms_df = pd.read_csv(rooms_db)

    # Kiểm tra xem Mã phòng có hợp lệ không
    if room_id_value not in rooms_df['ID_apartment'].values:
        with output_cus:
            clear_output(wait=True)
            print(f"Phòng '{room_id_value}' không tồn tại.")
    else:
        # Thêm thông tin thuê phòng vào tệp CSV khách hàng
        customer_data = {'ID': customer_id_value, 'First_name': first_name_value, 'Last_name': last_name_value,
                         'Age': age_value, 'Gender': gender_value, 'ID_apartment': room_id_value}
        customer_df = pd.read_csv(cus_db)
        customer_df = customer_df.append(customer_data, ignore_index=True)
        customer_df.to_csv(cus_db, index=False)

        # Cập nhật trạng thái của phòng thành 'For-rent'
        rooms_df.loc[rooms_df['ID_apartment'] == room_id_value, 'Status'] = 'For-rent'

        # Tăng 'Num_of_rentals' cho phòng trong DataFrame phòng
        rooms_df.loc[rooms_df['ID_apartment'] == room_id_value, 'Num_of_rentals'] += 1
        rooms_df.to_csv(rooms_db, index=False)
        update_room_options()
        update_room_options_empty()
        # Hiển thị thông báo xác nhận thuê phòng
        with output_cus:
            clear_output(wait=True)
            print(f"Đã thuê phòng cho khách hàng {first_name_value} {last_name_value} (ID: {customer_id_value}).")
    # Xóa nội dung các widgets sau khi thuê phòng
    clear_customer_inputs()

# Hàm xử lý xác nhận trả phòng
def confirm_return(b):
    # Lấy mã phòng đã chọn
    room_id_value = return_id_a.value.strip() if return_id_a.value else ''

    # Đọc DataFrame từ tệp phòng
    rooms_df = pd.read_csv(rooms_db)

    # Kiểm tra xem Mã phòng có hợp lệ không
    if room_id_value not in rooms_df['ID_apartment'].values:
        with output_cus:
            clear_output(wait=True)
            print(f"Phòng '{room_id_value}' không tồn tại.")
    else:
        # Xóa thông tin khách hàng ra khỏi tệp CSV khách hàng
        customer_df = pd.read_csv(cus_db)
        customer_df = customer_df[customer_df['ID_apartment'] != room_id_value]
        customer_df.to_csv(cus_db, index=False)

        # Cập nhật trạng thái của phòng thành 'Empty'
        rooms_df.loc[rooms_df['ID_apartment'] == room_id_value, 'Status'] = 'Empty'
        rooms_df.to_csv(rooms_db, index=False)
        update_room_options()
        update_room_options_empty()
        # Hiển thị thông báo xác nhận trả phòng
        with output_cus:
            clear_output(wait=True)
            print(f"Đã trả phòng {room_id_value}.")
    # Xóa nội dung các widgets sau khi thuê phòng
    clear_customer_inputs()

# Hàm xử lý tìm kiếm khách hàng
def search_customer(b):
    customer_room_value = search_customer_room.value.strip()
    customer_name_value = search_customer_name.value.strip().lower()

    # Đọc DataFrame từ tệp khách hàng
    customer_df = pd.read_csv(cus_db)

    # Tìm kiếm theo ID phòng
    if customer_room_value:
        result_df = customer_df[customer_df['ID_apartment'] == customer_room_value]
    # Tìm kiếm theo tên khách hàng (FirstName + LastName hoặc Last Name)
    elif customer_name_value:
        result_df = customer_df[customer_df.apply(lambda row: customer_name_value in row['First_name'].lower() + ' ' + row['Last_name'].lower() or customer_name_value in row['Last_name'].lower(), axis=1)]
    else:
        with output_cus:
            clear_output(wait=True)
            print("Vui lòng nhập ID phòng hoặc tên khách hàng để tìm kiếm.")
            return

    columns_to_display = ['ID', 'First_name', 'Last_name', 'Age', 'Gender', 'ID_apartment']
    result_df = result_df[columns_to_display]
    label_item.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Kết quả tìm kiếm:</div>'
    # Hiển thị kết quả tìm kiếm
    with output_cus:
        clear_output(wait=True)
        if result_df.empty:
            print("Không có thông tin khách hàng nào khớp với yêu cầu.")
        else:
            display(label_item, result_df)

def update_room_options():
    global search_customer_room  # Đảm bảo rằng search_customer_room được sử dụng global

    # Đọc DataFrame từ tệp phòng
    rooms_df = pd.read_csv(rooms_db)

    # Lọc các phòng có trạng thái "For-rent"
    for_rent_rooms = rooms_df[rooms_df['Status'] == 'For-rent']

    # Cập nhật giá trị options của Dropdown widget
    search_customer_room.options = sorted(list(for_rent_rooms['ID_apartment']))
    return_id_a.options = sorted(list(for_rent_rooms['ID_apartment']))

def update_room_options_empty():
    global search_customer_room  # Đảm bảo rằng search_customer_room được sử dụng global

    # Đọc DataFrame từ tệp phòng
    rooms_df = pd.read_csv(rooms_db)

    # Lọc các phòng có trạng thái "For-rent"
    for_rent_rooms = rooms_df[rooms_df['Status'] == 'Empty']

    # Cập nhật giá trị options của Dropdown widget
    rent_id_a.options = sorted(list(for_rent_rooms['ID_apartment']))

def update_room_options_all():
    global search_customer_room  # Đảm bảo rằng search_customer_room được sử dụng global

    # Đọc DataFrame từ tệp phòng
    rooms_df = pd.read_csv(rooms_db)

    # Cập nhật giá trị options của Dropdown widget
    id_all_room.options = sorted(list(rooms_df['ID_apartment']))


# Nút mở thêm phòng
def add_room_open(b):
    label_item.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Thêm phòng mới:</div>'
    with output:
          clear_output(wait=True)
          display(label_item, id_a, area, floor, price, upload_button, add_button)

# Nút mở cập nhật
def update_room_open(b):
    id_a_value = id_all_room.value.strip() if id_all_room.value else ''

    # Kiểm tra xem Mã phòng có hợp lệ không
    rooms_df = pd.read_csv(rooms_db)
    if id_a_value not in rooms_df['ID_apartment'].values:
        with output_cus:
            clear_output(wait=True)
            print(f"Phòng '{id_a_value}' không tồn tại.")
    else:
        # Lấy thông tin phòng từ DataFrame
        room_info = rooms_df.loc[rooms_df['ID_apartment'] == id_a_value].iloc[0]

        # Hiển thị các trường cập nhật với dữ liệu từ file CSV
        label_item.value = f'<div style="font-size: 14px; color: red; font-weight: bold;">Cập nhật phòng: {id_a_value}</div>'
        with output:
            clear_output(wait=True)
            update_area.value = str(room_info['Area'])
            update_floor.value = str(room_info['Floor'])
            update_price.value = str(room_info['Price'])
            display(label_item, update_area, update_floor, update_price, update_button)
        clear_inputs()

# Nút mở xoá phòng
def delete_room_open(b):
    label_item.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Xoá phòng:</div>'
    with output:
          clear_output(wait=True)
          display(label_item, delete_id_a, delete_button)

# Nút mở thêm khách hàng
def add_customer_layout(b):
    label_item.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Thêm khách hàng mới:</div>'
    with output_cus:
        clear_output(wait=True)
        display(label_item, customer_layout)



# Hàm thêm phòng
def add_room(b):
    id_a_value = id_a.value.strip() if id_a.value else ''
    area_value = area.value.strip() if area.value else ''
    floor_value = floor.value.strip() if floor.value else ''
    price_value = price.value.strip() if price.value else ''
    nor_value = 0

    # Đọc dữ liệu từ file CSV
    try:
        global rooms_df
    except pd.errors.EmptyDataError:
        # Xử lý trường hợp nếu file CSV trống
        rooms_df = pd.DataFrame(columns=headers)

    # Kiểm tra xem phòng đã tồn tại chưa
    if id_a_value in rooms_df['ID_apartment'].values:
        print(f"Phòng '{id_a_value}' đã tồn tại. Sử dụng chức năng cập nhật để thay đổi.")
    else:
        # Thêm phòng mới vào DataFrame
        new_room = {'ID_apartment': id_a_value, 'Area': area_value, 'Floor': floor_value, 'Price': price_value, 'Status': "Empty", 'Num_of_rentals': nor_value}
        rooms_df = rooms_df.append(new_room, ignore_index=True)

        # Lưu DataFrame vào file CSV
        rooms_df.to_csv(rooms_db, index=False)
        upload_images(id_a_value)
        update_room_options()
        update_room_options_all()
        update_room_options_empty()
        with output:
            clear_output(wait=True)
            print(f"Phòng '{id_a_value}' đã được thêm thành công.")
    # Xóa nội dung các widgets sau khi thêm phòng
    clear_inputs()

def upload_images(id_a_value):
    uploaded_files = upload_button.value

    if not uploaded_files:
        with output:
            clear_output(wait=True)
            print("Vui lòng chọn ít nhất một tệp tin hình ảnh.")
        return

    attachment_df = pd.read_csv(csv_pathAttachment)

    for filename, file_info in uploaded_files.items():
        image_path = f'/content/drive/MyDrive/CSV/Images/{filename}'
        attachment_data = {'ID_attachment': len(attachment_df) + 1, 'Image_Path': image_path, 'ID_apartment': id_a_value}
        attachment_df = attachment_df.append(attachment_data, ignore_index=True)

        with open(image_path, 'wb') as f:
            f.write(file_info['content'])

    # Lưu DataFrame vào file CSV
    attachment_df.to_csv(csv_pathAttachment, index=False)

def show_room_detail(b):
    global rooms_df
    id_a_detail_value = id_all_room.value.strip()

    attachment_df = pd.read_csv(attachment_db)

    room_info = rooms_df[rooms_df['ID_apartment'] == id_a_detail_value]

    has_room_info = not room_info.empty

    # Hiển thị thông tin phòng
    with output:
        clear_output(wait=True)
        if not has_room_info:
            print(f"Không có thông tin cho ID phòng '{id_a_detail_value}'.")
        else:

            # Lọc attachment_df theo ID_apartment
            apartment_attachments = attachment_df[attachment_df['ID_apartment'] == id_a_detail_value]

            if apartment_attachments.empty:
                display(room_info)
                print(f"Không có hình ảnh cho ID phòng '{id_a_detail_value}'.")
            else:
                num_images = len(apartment_attachments)
                num_cols = min(3, num_images)
                num_rows = (num_images // num_cols) + 1
                fig, axes = plt.subplots(num_rows, num_cols + 1, figsize=(15, 4 * num_rows))

                room_info_text = f"Thông tin phòng {id_a_detail_value}:\n\n"
                for idx, (column, value) in enumerate(room_info.iloc[0].items()):
                    room_info_text += f"{column}: {value}\n"
                axes[0, -1].text(0, 0.5, room_info_text, fontsize=12, va='center', ha='left')
                axes[0, -1].axis('off')

                for i, (index, row) in enumerate(apartment_attachments.iterrows()):
                    image_path = row['Image_Path']

                    img = PILImage.open(image_path)
                    img = img.resize((150, 150))


                    row_idx = i // num_cols
                    col_idx = i % num_cols


                    axes[row_idx, col_idx].imshow(img)
                    axes[row_idx, col_idx].axis('off')
                    axes[row_idx, col_idx].set_title(f'Ảnh {i + 1}')

                plt.show()


# Hàm cập nhật trạng thái phòng
def update_room(b):
    id_a_value = id_all_room.value
    area_value = update_area.value
    floor_value = update_floor.value
    price_value = update_price.value
    rooms_df = pd.read_csv(rooms_db)

    if id_a_value not in rooms_df['ID_apartment'].values:
        print(f"Phòng '{id_a_value}' không tồn tại. Sử dụng chức năng thêm phòng để thêm mới.")
    else:
        rooms_df.loc[rooms_df['ID_apartment'] == id_a_value, 'Area'] = area_value
        rooms_df.loc[rooms_df['ID_apartment'] == id_a_value, 'Floor'] = floor_value
        rooms_df.loc[rooms_df['ID_apartment'] == id_a_value, 'Price'] = price_value
        rooms_df.to_csv(rooms_db, index=False)
        with output:
              clear_output(wait=True)
              print(f"Trạng thái của phòng '{id_a_value}' đã được cập nhật thành công.")

# Hàm xóa phòng
def delete_room(b):
    id_a_value = delete_id_a.value
    rooms_df = pd.read_csv(rooms_db)
    attachment_df = pd.read_csv(attachment_db)
    customer_df = pd.read_csv(cus_db)
    if id_a_value not in rooms_df['ID_apartment'].values:
        with output:
            clear_output(wait=True)
            print(f"Phòng '{id_a_value}' không tồn tại.")
    else:
        if id_a_value in for_rent_rooms['ID_apartment'].values or id_a_value in attachment_df['ID_apartment'].values:
          label_item.value = f'<div style="font-size: 14px; color: red; font-weight: bold;">Xoá phòng:{id_a_value}</div>'
          label_item2.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Phòng có người thuê</div>'
          with output:
            clear_output(wait=True)
            display(label_item, label_item2, delete_button_rent, delete_button_cancel )

        else:
          rooms_df = rooms_df[rooms_df['ID_apartment'] != id_a_value]
          rooms_df.to_csv(rooms_db, index=False)
          update_room_options_empty()
          update_room_options_all()
          with output:
              clear_output(wait=True)
              print(f"Phòng '{id_a_value}' đã được xóa thành công.")

def delete_room_rent(b):
    id_a_value = delete_id_a.value
    rooms_df = pd.read_csv(rooms_db)
    rooms_df = rooms_df[rooms_df['ID_apartment'] != id_a_value]
    rooms_df.to_csv(rooms_db, index=False)
    update_room_options_empty()
    update_room_options_all()
    customer_df = pd.read_csv(cus_db)
    customer_df = customer_df[customer_df['ID_apartment'] != id_a_value]
    customer_df.to_csv(cus_db, index=False)
    attachment_df = pd.read_csv(attachment_db)
    attachment_df = attachment_df[attachment_df['ID_apartment'] != id_a_value]
    attachment_df.to_csv(attachment_db, index=False)
    with output:
          clear_output(wait=True)
          print(f"Phòng '{id_a_value}' đã được xóa thành công.")

def list_room(b):
    list_rooms = pd.read_csv(rooms_db)

    # Kiểm tra và xử lý NaN nếu có
    if list_rooms.isna().any().any():
        list_rooms = list_rooms.dropna()
    list_rooms['ID_apartment'] = list_rooms['ID_apartment']
    list_rooms['Floor'] = list_rooms['Floor'].astype(int)
    list_rooms['Num_of_rentals'] = list_rooms['Num_of_rentals'].astype(int)
    list_rooms = list_rooms.sort_values(by='ID_apartment', ascending=True)
    label_item.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Danh sách phòng</div>'
    with output:
        clear_output(wait=True)
        display(label_item, list_rooms)

# Hàm xử lý liệt kê phòng trống
def list_empty_rooms(b):
    rooms_df = pd.read_csv(rooms_db)
    empty_rooms = rooms_df[rooms_df['Status'] == 'Empty']
    label_item.value = '<div style="font-size: 14px; color: red; font-weight: bold;">Danh sách phòng trống:</div>'
    with output:
        clear_output(wait=True)
        if empty_rooms.empty:
            print("Không có phòng trống.")
        else:
            display(label_item, empty_rooms)

# Hàm để hiển thị biểu đồ 10 phòng được thuê nhiều nhất
def on_chart_button_click(b):
    with output:
        clear_output(wait=True)
        show_top_rent_rooms()


# Hàm hiển thị biểu đồ cho 10 phòng được thuê nhiều nhất
def show_top_rent_rooms():
    rooms_df = pd.read_csv(rooms_db)


    sorted_rooms = rooms_df.sort_values(by='Num_of_rentals', ascending=False)


    top_10_rooms = sorted_rooms.head(10)


    total_rentals = top_10_rooms['Num_of_rentals'].sum()


    top_10_rooms['Percentage'] = (top_10_rooms['Num_of_rentals'] / total_rentals) * 100


    ids = top_10_rooms['ID_apartment'].astype(str)
    percentages = top_10_rooms['Percentage']

    # Vẽ biểu đồ cột
    plt.figure(figsize=(10, 6))
    bars = plt.bar(ids, percentages, color='#7F8F9C')
    plt.xlabel('ID phòng')
    plt.ylabel('Phần trăm số Lượt thuê (%)')
    plt.title('10 phòng thuê nhiều nhất (theo phần trăm)')
    plt.xticks(rotation=45, ha="right")


    for bar, percentage in zip(bars, percentages):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 1, f'{percentage:.2f}%', ha='center', va='bottom')

    plt.tight_layout()

    # Hiển thị biểu đồ
    plt.show()

# Hàm để xóa nội dung của các widgets sau khi thêm, sửa, xoá sách
def clear_inputs():
    id_a.value = ''
    area.value = ''
    floor.value = ''
    price.value = ''
    customer_id.value = ''
    gender.value = ''
    update_price.value = ''
    update_area.value = ''
    update_floor.value = ''
    delete_id_a.value = ''
    return_id_a.value = ''
    rent_id_a.value = ''
    first_name.value = ''
    last_name.value = ''
    age.value = ''
    search_customer_name.value = ''

def clear_customer_inputs():
    customer_id.value = ''
    first_name.value = ''
    last_name.value = ''
    search_customer_room.value = ''

# Kết nối các hàm xử lý sự kiện với các nút
# chart_button.on_click(lambda b: show_top_rent_rooms())
chart_button.on_click(on_chart_button_click)
rent_button_confirm.on_click(confirm_rent)
add_button.on_click(add_room)
add_button_open.on_click(add_room_open)
detail_button.on_click(show_room_detail)
update_button.on_click(update_room)
update_button_open.on_click(update_room_open)
delete_button.on_click(delete_room)
delete_button_rent.on_click(delete_room_rent)
delete_button_open.on_click(delete_room_open)
delete_button_cancel.on_click(delete_room_open)
list_button.on_click(list_room)
list_empty_button.on_click(list_empty_rooms)
rent_button.on_click(rent_room)
return_button.on_click(confirm_return)
search_customer_button.on_click(search_customer)
add_customer_button.on_click(confirm_add_customer)
add_customer_button_open.on_click(add_customer_layout)
status_button_pie_chart.on_click(update_room_status_pie_chart)
list_rooms_by_floor_pie_chart_button.on_click(update_rooms_by_floor_pie_chart)
search_drop.observe(on_change)
search_button.on_click(handle_search_button_click)
floor_dropdown.observe(update_chart, names='value')

# Tạo layout
layout_search_drop = widgets.HBox([search_drop], layout=widgets.Layout(margin='10px'))
layout_room_update = widgets.HBox([id_all_room, detail_button], layout=widgets.Layout(margin='10px'))
button_layout_room1 = widgets.HBox([add_button_open, delete_button_open, update_button_open, list_button], layout=widgets.Layout(margin='10px'))
button_layout_room2 = widgets.HBox([list_empty_button, chart_button, status_button_pie_chart, list_rooms_by_floor_pie_chart_button], layout=widgets.Layout(margin='10px'))
button_layout_room3 = widgets.HBox([floor_dropdown], layout=widgets.Layout(margin='10px'))

search_customer_layout = widgets.HBox([search_customer_room, search_customer_name, search_customer_button],
                                      layout=widgets.Layout(flex_flow='column', margin='10px'))
button_layout_cus1 = widgets.HBox([rent_id_a, rent_button, add_customer_button_open, return_id_a, return_button], layout=widgets.Layout(flex_flow='column', margin='10px'))
customer_layout = widgets.VBox([customer_id, first_name, last_name, age, gender, search_customer_room, add_customer_button], layout=widgets.Layout(flex_flow='column', margin='10px'))
# Hiển thị widgets
display(label_room)
display(layout_search_drop)
display(output_search)
display(layout_room_update)
display(button_layout_room1)
display(button_layout_room2)
display(button_layout_room3)
display(output)
display(label_cus)
display(search_customer_layout)
display(button_layout_cus1)
display(output_cus)
update_chart({'new': floor_options[0]})
```
